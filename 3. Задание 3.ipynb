```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Задание 3

Предположим, каждый день вы получаете предсказание   f_i величины r_i

Введем понятие Profit-and-loss (pnl) для каждого дня

pnl_i  = f_i   * r_i

Пусть f_i имеет распределение standard Gaussian с mean = 0 и std = 1. 

Задача: Сгенерируйте несколько случайных реализаций f_i, постройте графики pnl_i  и pnl = cumsum(pnl_i)


```python
f_1 = np.random.standard_normal(100)
f_2 = np.random.standard_normal(1000)
f_3 = np.random.standard_normal(10000)
f_4 = np.random.standard_normal(100000)
```

```python
fig, ax = plt.subplots(2,2, figsize=(20,10))
fig.suptitle('Выборки со standard Gaussian распределением (mean = 0 и std = 1)', fontsize=20)
sns.histplot(x=f_1, ax=ax[0][0], label='100')
sns.histplot(x=f_2, ax=ax[0][1], label='1_000')
sns.histplot(x=f_3, ax=ax[1][0], label='10_000')
sns.histplot(x=f_4, ax=ax[1][1], label='100_000')

ax[0][0].legend(loc='upper right', fontsize=15)
ax[0][1].legend(loc='upper right', fontsize=15)
ax[1][0].legend(loc='upper right', fontsize=15)
ax[1][1].legend(loc='upper right', fontsize=15)
plt.show()
```

Так как не сказано какая величина r_i, но указано, что получаем предсказание f_i величины r_i, то по логике r_i должна иметь такое же распределение, что и f_i. 

```python
r_1 = np.random.standard_normal(100)
r_2 = np.random.standard_normal(1000)
r_3 = np.random.standard_normal(10000)
r_4 = np.random.standard_normal(100000)
```

```python
pnl_1 = f_1 * r_1
pnl_2 = f_2 * r_2
pnl_3 = f_3 * r_3
pnl_4 = f_4 * r_4
```

```python
pnl_cs1 = np.cumsum(pnl_1)
pnl_cs2 = np.cumsum(pnl_2)
pnl_cs3 = np.cumsum(pnl_3)
pnl_cs4 = np.cumsum(pnl_4)
```

```python
plt.figure(figsize=(20,5))
plt.title('Распределение во времени  pnl_i и pnl по выборке 100', fontsize=20)
plt.plot(pnl_1,  label='pnl_1')
plt.plot(pnl_cs1, label='pnl_cs1')
plt.legend(loc='lower left', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.title('Распределение во времени  pnl_i и pnl  по выборке 1_000', fontsize=20)
plt.plot(pnl_2,  label='pnl_2')
plt.plot(pnl_cs2, label='pnl_cs2')
plt.legend(loc='lower left', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.title('Распределение во времени  pnl_i и pnl  по выборке 10_000', fontsize=20)
plt.plot(pnl_3,  label='pnl_3')
plt.plot(pnl_cs3, label='pnl_cs3')
plt.legend(loc='lower left', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.title('Распределение во времени  pnl_i и pnl  по выборке 100_000', fontsize=20)
plt.plot(pnl_4,  label='pnl_4')
plt.plot(pnl_cs4, label='pnl_cs4')
plt.legend(loc='lower left', fontsize=15)
plt.show()
```
