```python
# Загружаем данные

google = pd.read_csv('data/google_c.csv')
```

```python
def data_preparation(df):
    """Подготовка данных для анализа:
       индексы - даты, колонка - закрытие торгов
       данные за 10 лет"""
    
    df = df.rename(columns={'4. close':'close', 'Unnamed: 0':'date'})
    df['date'] = pd.to_datetime(df['date'])
    df = df[df['date']>'2012-08-23'] # по 23 - чтоб потом удалить одну строку
    df = df.set_index(pd.DatetimeIndex(df['date']))
    df = df.sort_index() 
    
    return df[['close']]   
```

```python
google = data_preparation(google)
```

```python
google['close_next'] = google['close'].shift(-1)
google = google.dropna(axis=0)
google['daily_returns'] = (google['close_next'] - google['close'])/google['close']
google
```

```python
# Статистические показатели

google['daily_returns'].describe()
```

```python
fig, ax = plt.subplots(2,1, figsize=(20,7))
fig.suptitle('Распределение daily_returns по акциям Google', fontsize=20)
sns.histplot(x=google['daily_returns'], ax=ax[0])
ax[0].set_xlabel('')
sns.boxplot(x=google['daily_returns'], ax=ax[1])
plt.show()
```

В статистических показателях и на графиках присутствуют аномально большие значения. По графику boxplot видно, что оно одно.

Посмотрим как выглядят графики, если отсечь это значение.

```python
fig, ax = plt.subplots(2,1, figsize=(20,7))
fig.suptitle('Распределение daily_returns по акциям Google (без аномально высоких значений)', fontsize=20)
sns.histplot(x=google[google['daily_returns']>-0.2]['daily_returns'], ax=ax[0])
ax[0].set_xlabel('')
sns.boxplot(x=google[google['daily_returns']>-0.2]['daily_returns'], ax=ax[1])
plt.show()
```

**Проверна на нормальное гаусовское распределение**

Визуально распределение похоже на нормальное (гаусовское). Но нужно проверить.

```python
# Аномальные значенния

google[google['daily_returns']<-0.2]
```

**Критерии нормального распределения:**
     
     - распределение симетрично и имеет форму колокола (если не считать 1 аномалии).
     - среднее значение (0.000362) и медиана (0.000950) практически совпадают - под вопросом.
     - В пределах одного стандартного отклонения в обе стороны лежат 68.2% всех данных, в пределах двух - 95,5%, в пределах трех - 99,7% - это надо проверять:

```python
m = google['daily_returns'].mean()
s = google['daily_returns'].std()
min_1s = m-s
max_1s = m+s
print('Среднее %.3f, станд. откл. %.3f,  сред-1стд %.3f,  сред+1стд %.3f' %(m, s, min_1s, max_1s))
google[(google['daily_returns']>min_1s) & (google['daily_returns']<max_1s)][['daily_returns']].agg(['min', 'max', 'count'])
```

```python
# Процент наблюдений, содержащийся в пределах одного стандартного отклонения 
# у нас 91%  наблюдений 

1924/2120
```

Проведем стат тесты:

```python
print('тест Шапиро-Уилк')
stat, p = scipy.stats.shapiro(google['daily_returns']) 
print('Statistics=%.3f, p-value=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

```python
print('Критерий согласия Пирсона')
stat, p = scipy.stats.normaltest(google['daily_returns']) 
print('Statistics=%.3f, p-value=%.3f' % (stat, p)) 
if p > alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

Вывод: Данное распределение по тесстам не является нормальным.

**Проверка на стационарность**

```python
plt.figure(figsize=(20,5))
plt.title('Распределение daily_returns во времени', fontsize=20)
plt.plot(google['daily_returns'])
plt.xlabel('Период', fontsize=15)
plt.ylabel('daily_returns', fontsize=15)
plt.show()
```

```python
print('Тест Дикки-Фуллера')
test = sm.tsa.adfuller(google['daily_returns'])
print('p-value:', test[1])
if test[1]<0.05:
    print('ряд стационарный')
else:
    print('ряд не стационарный')
```

По тесту и зрительно ряд выглядит стационарным. Имеются одно аномально низкое значение.
